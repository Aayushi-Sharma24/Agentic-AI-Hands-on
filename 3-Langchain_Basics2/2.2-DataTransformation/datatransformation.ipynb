### RecursiveTextSplitter

```python
## Reading a PDf File
from langchain_community.document_loaders import PyPDFLoader
loader=PyPDFLoader('syllabus.pdf')
docs=loader.load()
docs
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

## using split documents -- used when data is already converted into docs
text_splitter=RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=50)
final_docs=text_splitter.split_documents(docs)
final_docs
```

```python
## using create documents-- used cz speech is in the form of text/string

speech = ""
with open("speech.txt") as f:
    speech = f.read()

text_splitter = RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=50)
text = text_splitter.create_documents([speech])
text
```

### CharacterTextSplitter

```python
from langchain_text_splitters import CharacterTextSplitter

#used split documents
text_splitter=CharacterTextSplitter(separator="\n\n", chunk_size=1000, chunk_overlap=50)
text = text_splitter.split_documents(docs)
text
```

```python
## using create documents
speech=""
with open("speech.txt") as f:
    speech=f.read()


text_splitter=CharacterTextSplitter(chunk_size=100,chunk_overlap=20)
text=text_splitter.create_documents([speech])
text
```

### HTMLHeaderTextSplitter

```python
from langchain_text_splitters import HTMLHeaderTextSplitter

html_string = """
<!DOCTYPE html>
<html>
<body>
    <div>
        <h1>Foo</h1>
        <p>Some intro text about Foo.</p>
        <div>
            <h2>Bar main section</h2>
            <p>Some intro text about Bar.</p>
            <h3>Bar subsection 1</h3>
            <p>Some text about the first subtopic of Bar.</p>
            <h3>Bar subsection 2</h3>
            <p>Some text about the second subtopic of Bar.</p>
        </div>
        <div>
            <h2>Baz</h2>
            <p>Some text about Baz</p>
        </div>
        <br>
        <p>Some concluding text about Foo</p>
    </div>
</body>
</html>
"""

headers_to_split_on= [
    ("h1","Header 1"),
    ("h2","Header 2"),
    ("h3","Header 3")
]

html_splitter=HTMLHeaderTextSplitter(headers_to_split_on)
html = html_splitter.split_text(html_string)
html
```

### JsonSplitter

```python
import json
import requests
from langchain_text_splitters import RecursiveJsonSplitter

json_data=requests.get("https://api.smith.langchain.com/openapi.json").json()

json_splitter=RecursiveJsonSplitter(max_chunk_size=300)
json_chunks=json_splitter.split_json(json_data)
json_chunks
```
