## Embedding Techniques
### Converting text to vectors

```python
import os
from dotenv import load_dotenv

load_dotenv()
os.environ["OPENAI_API_KEY"] = os.getenv("OPENAI_API_KEY")
```

```python
from langchain_openai import OpenAIEmbeddings

embeddings=OpenAIEmbeddings(model="text-embedding-3-large")
embeddings
```

```python
text = "This is sample text used for creating OpenAI Embeddings"
embed_query=embeddings.embed_query(text)
embed_query
```

```python
len(embed_query)
```

### Using dimensions to reduce dimensionality

```python
from langchain_openai import OpenAIEmbeddings

embeddings = OpenAIEmbeddings(model="text-embedding-3-large",dimensions=1024)
text_query = "Hello this is sample text"
result = embeddings.embed_query(text_query)
print(result)
print("embedings len",len(result))
```

### Embedding documents

```python
from langchain_community.document_loaders import TextLoader

loader=TextLoader("speech.txt")
docs=loader.load()
docs
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

splitter=RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=100)
final_docs=splitter.split_documents(docs)
final_docs
```

```python
final_docs[0].page_content
```

```python
for doc in final_docs:
    document = doc.page_content
    print(document)
```

```python
from langchain_openai import OpenAIEmbeddings
embeddings = []

openai_embeddings=OpenAIEmbeddings(model="text-embedding-3-large")
for doc in final_docs:
    document = doc.page_content
    embedding=openai_embeddings.embed_query(document)
    embeddings.append(embedding)

embeddings
```

### Batch Embedding for Speed if no of documents are large

```python
texts = [doc.page_content for doc in final_docs]
embeddings_batch= openai_embeddings.embed_documents(texts)
embeddings_batch
```

### HuggingFace Embedding

```python
## used finegrained token
from langchain_huggingface import HuggingFaceEmbeddings

embeddings=HuggingFaceEmbeddings(model_name="all-MiniLM-L6-v2")
embeddings
```
